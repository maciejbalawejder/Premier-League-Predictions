```python
import pandas as pd 
import numpy as np 
from ELO import ELO_rating
```

```python
season19 = pd.read_csv('PL19-20.csv')
season19['AwayTeam'] = season19['AwayTeam'].replace(['Brighton','Leeds','Leicester','Man City','Man United','Newcastle','Tottenham','West Brom','West Ham','Wolves','Norwich'],['Brighton & Hove Albion','Leeds United','Leicester City','Manchester City','Manchester United','Newcastle United','Tottenham Hotspur','West Bromwich Albion','West Ham United','Wolverhampton Wanderers','Norwich City'])
season19['HomeTeam'] = season19['HomeTeam'].replace(['Brighton','Leeds','Leicester','Man City','Man United','Newcastle','Tottenham','West Brom','West Ham','Wolves','Norwich'],['Brighton & Hove Albion','Leeds United','Leicester City','Manchester City','Manchester United','Newcastle United','Tottenham Hotspur','West Bromwich Albion','West Ham United','Wolverhampton Wanderers','Norwich City'])
teams = season19['HomeTeam'].unique()
teams = sorted(teams)
print(teams)
```

```python
playing_style = pd.read_csv('FIFA21_data/teams.csv')
playing_style['str_defensive_style'], _ = pd.factorize(playing_style['str_defensive_style'])
playing_style['str_offensive_style'], _ = pd.factorize(playing_style['str_offensive_style'])
idxs = [10,13]
cols = playing_style.columns
style_data = {}
for team in teams:
    style_data[team] = np.array(playing_style[playing_style['str_team_name'] == team][cols[idxs]])
style_data['Arsenal'][0]
```

```python
dates = {'Arsenal': ['23:05:2018', '23:12:2019'], 'Aston Villa': ['03:10:2018'], 'Brighton & Hove Albion': ['20:05:2019'], 'Burnley': ['30:10:2012'], 'Chelsea': ['04:07:2019', '26:01:2021'], 'Crystal Palace': ['12:09:2017'], 'Everton': ['31:05:2018', '23:12:2019'], 'Fulham': ['28:02:2019'], 'Leeds United': ['15:01:2018'], 'Leicester City': ['27:02:2019'], 'Liverpool': ['08:10:2015'], 'Manchester City': ['01:07:2016'], 'Manchester United': ['19:12:2018'], 'Newcastle United': ['17:07:2019'], 'Sheffield United': ['12:05:2016', '13:03:2021'], 'Southampton': ['06:12:2018'], 'Tottenham Hotspur': ['14:05:2014','20:11:2019', '19:04:2021'], 'West Bromwich Albion': ['13:01:2019', '16:12:2020', '24:01:2021'], 'West Ham United': ['22:05:2018', '30:12:2019'], 'Wolverhampton Wanderers': ['31:05:2017']}
s19managers = {}
missing = []
for team in teams:
    try :
        s19managers[team] = dates[team]
    except:
        missing.append(team)
    
s19managers[missing[0]] = ['12:10:2012']
s19managers[missing[1]] = ['21:01:2018', '07:09:2019','06:12:2019']
s19managers[missing[2]] = ['01:07:2017']
print(s19managers)
```

```python
import os
import re
import time 
path = 'Fantasy-Premier-League/data/2019-20/'
understats_columns = ['xG','xGA','ppda','date']
team_understats = {}
files = sorted(os.listdir(path+'understat/'))[1:]
for idx,file in enumerate(files):
    row = pd.read_csv(path+'understat/'+file)[understats_columns]
    for loc in range(row.shape[0]):
        nums = re.findall(r'\d+', row['ppda'].loc[loc])
        res = int(nums[0]) / int(nums[1])
        row['ppda'].loc[loc] = res
    team_understats[teams[idx]] = row
team_understats
```

```python
ratings_columns = ['name','strength_overall_home', 'strength_overall_away']
ratings = pd.read_csv(path+'teams.csv')[ratings_columns]
df = ratings.transpose()
df = df.rename(columns=df.iloc[0])
df = df.iloc[1:,:]
ratings = df
ratings.columns = teams
ratings
```

```python
'''
AS - attacking style 
DS - defensive style 
SOH - strength overall home
SOA - strength overall away 
xG - quality of goal scoring chances 
xGa - quality of preventing scoring chances 
ppda - passes allowed per defensive action 
Mtime - managers cadance time
B365H - odds for home team
B365D - odds for draw
B365A - odds for away team
'''
cols = ['Date','HomeTeam', 'AwayTeam','B365H','B365D','B365A', 'FTHG', 'FTAG', 'FTR']
s19data = season19[cols]
input_data = pd.DataFrame(columns= ['HAS', 'HDS', 'HSOH', 'HxG', 'HxGa', 'Hppda', 'HMtime',
                                    'AAS', 'ADS', 'ASOH', 'AxG', 'AxGa', 'Appda', 'AMtime',
                                    'B365H', 'B365D', 'B365A','FTR','HomeTeam', 'AwayTeam'])
def find_xG(team, date):
    for _, row in team_understats[team].iterrows():
        d = row['date']
        d = sorted(d[:10].split('-'))
        if d == sorted(date.split('/')):
            xG = row['xG']
            xGa = row['xGA']
            ppda = row['ppda']
            break

    return xG, xGa, ppda    

def managerTime(match_date, manager_date):
    match_time = time.mktime(time.strptime(match_date, "%d/%m/%Y"))
    managers_time = [time.mktime(time.strptime(period.replace(':','/'), "%d/%m/%Y")) - match_time for period in manager_date if time.mktime(time.strptime(period.replace(':','/'), "%d/%m/%Y")) - match_time < 0]
    managers_time = max(managers_time) / 86_400
    return -managers_time
 
    
for index,match in s19data.iterrows():
    row = []
    date = match[0]
    homeTeam = match[1]
    awayTeam = match[2]
    oddsHome = match[3]
    oddsDraw = match[4]
    oddsAway = match[5]
    homeScore = match[6]
    awayScore = match[7]
    FTR = match[8]
    HAS = style_data[homeTeam][0][1]
    HDS = style_data[homeTeam][0][0]
    AAS = style_data[awayTeam][0][1]
    ADS = style_data[awayTeam][0][0]
    
    homeRaiting, awayRaiting = ratings[homeTeam].iloc[0], ratings[awayTeam].iloc[1]
    
    HSOH = ELO_rating(homeRaiting,awayRaiting,homeScore,awayScore,'h')
    ASOH = ELO_rating(homeRaiting,awayRaiting,homeScore,awayScore,'a')
    
    ratings[homeTeam].iloc[0], ratings[awayTeam].iloc[1] = HSOH, ASOH
    
    HxG, HxGa, Hppda = find_xG(homeTeam,date)
    AxG, AxGa, Appda = find_xG(awayTeam,date)
    
    home_manager, away_manager = s19managers[homeTeam], s19managers[awayTeam]
    HMtime, AMtime = managerTime(date,home_manager),managerTime(date,away_manager)
    
    row = [HAS,HDS,HSOH,HxG,HxGa,Hppda,HMtime,AAS,ADS,ASOH,AxG,AxGa,Appda,AMtime,oddsHome,oddsDraw,oddsAway,FTR,homeTeam,awayTeam]
    
    input_data.loc[index] = row    

    
    
    
                
    
    

input_data.to_csv('19-20data.csv', index = False)    

print(input_data['HSOH'])
```

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.preprocessing import MinMaxScaler
```

```python
inputs = pd.read_csv('19-20data.csv')
inputs.head()
scaler = MinMaxScaler()
inputs[inputs.columns[:-3]] = scaler.fit_transform(inputs[inputs.columns[:-3]])
inputs.head()
```

```python
def sequence_data(X,timestep = 5):
    # dims of data [batch,timestep,feature] in our case it's harder cause we need to prepare 5 last matches for both 
    teams_name = sorted(X['HomeTeam'].unique())
    teams_results = {}
    teams_data = {}
    final = {}
    Y = {}
    for name in teams_name:
        teams_data[name] = []
        teams_results[name] = []
        final[name] = []
        Y[name] = []
    for idx, match in X.iterrows():
        homeTeam = match[-2]
        awayTeam = match[-1]
        
        teams_data[homeTeam].append(match[:-2])
        teams_data[awayTeam].append(match[:-2])

        
    for team,matches in teams_data.items():
        for idx in range(len(matches)-timestep):
            final[team].append(matches[idx:idx+timestep])
    return final
        
        
            
x_seq = sequence_data(inputs)
```

```python
input_data = np.empty((0,5,18))
for key,val in x_seq.items():
    input_data = np.vstack((input_data,np.asarray(val)))
y = []
for idx,sample in enumerate(input_data):
    for jdx,timestep in enumerate(sample):
        if jdx == input_data.shape[1] - 1:
            if timestep[-1] == 'A':
                y.append([1,0,0])
            elif timestep[-1] == 'D':
                y.append([0,1,0])
            elif timestep[-1] == 'H':
                y.append([0,0,1])
                
y = np.asarray(y)
input_data = np.delete(arr = input_data, obj = 17, axis = 2)
print(input_data.shape, y.shape)
```

```python
np.save('inputs19:20.npy',input_data)
np.save('outputs19:20.npy',y)
```
