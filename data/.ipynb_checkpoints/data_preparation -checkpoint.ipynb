```python
import pandas as pd
import os
import numpy as np
import re
import matplotlib.pyplot as plt
folders = os.listdir()
print(folders)
from ELO import ELO_rating
import time
```

```python
def teams(folders):
    tactics_data = pd.read_csv("{}/tbl_team_tactics.csv".format(folders[0]))
    ratings_data = pd.read_csv("{}/tbl_team.csv".format(folders[0]))
    teams_data = pd.merge(ratings_data, tactics_data)
    teams_data = teams_data[teams_data['str_league'] == 'English Premier League (1)']
    cols = teams_data.columns
    save = [1,11,14]
    teams_data = teams_data.drop(columns = [cols[i] for i in range(len(cols)) if i not in save])
    columns_name = teams_data.columns

    teams_data[columns_name[-2]], def_style = pd.factorize(teams_data[columns_name[-2]])
    teams_data[columns_name[-1]], att_style = pd.factorize(teams_data[columns_name[-1]])
    teams_data = teams_data.sort_values(by = 'str_team_name')
    teams_data = teams_data.reset_index(drop=True)
    
    print(def_style)
    print(att_style)
    
    return teams_data
teams_data = teams(folders)
datas = os.listdir("{}/data/2020-21/".format(folders[3]))
teams_ratings = pd.read_csv(r'{}/data/2020-21/{}'.format(folders[3],datas[1]))
save = ['name', 'strength_overall_home', 'strength_overall_away']
teams_ratings = teams_ratings[save].sort_values(by = 'name').drop(columns = 'name')
teams_data = pd.concat([teams_data,teams_ratings], axis = 1)
teams_data.to_csv('teams_data.csv')
teams_data.T
```

```python
df = teams_data.transpose()
df = df.rename(columns=df.iloc[0])
df = df.iloc[1:,:]
df.to_csv('Team_style.csv')
teams_data = df
```

```python
def games(folders):
    matches = os.listdir(r'{}/data/2020-21/{}'.format(folders[3],datas[2])) # each dataset of matches for each team
    matches.remove('understat_player.csv')
    cols_to_save = ['xG','xGA','ppda','date']
    
    for idx in range(len(matches)):
        datav0 = pd.read_csv(r'{}/data/2020-21/{}/{}'.format(folders[3],datas[2],matches[idx]))[cols_to_save]
        for loc in range(datav0.shape[0]):
            nums = re.findall(r'\d+', datav0['ppda'].loc[loc])
            res = int(nums[0]) / int(nums[1])
            datav0['ppda'].loc[loc] = res
        datav0.to_csv(r"teams_matches/{}".format(matches[idx][10:]))
df = games(folders)
```

```python
# row contains [data, hometeam, awayteam, OddsHome, OddsDraw, OddsAway]
matches['AwayTeam'] = matches['AwayTeam'].replace(['Brighton','Leeds','Leicester','Man City','Man United','Newcastle','Tottenham','West Brom','West Ham','Wolves'],['Brighton & Hove Albion','Leeds United','Leicester City','Manchester City','Manchester United','Newcastle United','Tottenham Hotspur','West Bromwich Albion','West Ham United','Wolverhampton Wanderers'])
matches['HomeTeam'] = matches['HomeTeam'].replace(['Brighton','Leeds','Leicester','Man City','Man United','Newcastle','Tottenham','West Brom','West Ham','Wolves'],['Brighton & Hove Albion','Leeds United','Leicester City','Manchester City','Manchester United','Newcastle United','Tottenham Hotspur','West Bromwich Albion','West Ham United','Wolverhampton Wanderers'])

style_data = pd.read_csv(r"Team_style.csv")
input_data = pd.DataFrame(columns= ['HAS', 'HDS', 'HSOH', 'HxG', 'HxGa', 'Hppda', 'HMtime',
                                    'AAS', 'ADS', 'ASOH', 'AxG', 'AxGa', 'Appda', 'AMtime',
                                    'B365H', 'B365D', 'B365A','FTR','HomeTeam', 'AwayTeam'])

def find_xG(team, date):
    for _, row in teams[team].iterrows():
        d = row['date']
        d = sorted(d[:10].split('-'))
        if d == sorted(date.split('/')):
            xG = row['xG']
            xGa = row['xGA']
            ppda = row['ppda']
            break

    return xG, xGa, ppda    

def managerTime(match_date, manager_date):
    match_time = time.mktime(time.strptime(match_date, "%d/%m/%Y"))
    manager_date = manager_date[0].replace("'","").replace('[',"").replace("]","").replace(":","/").replace(' ','').split(',')
    managers_time = [time.mktime(time.strptime(period, "%d/%m/%Y")) - match_time for period in manager_date if time.mktime(time.strptime(period, "%d/%m/%Y")) - match_time < 0]
    managers_time = max(managers_time) / 86_400
    return -managers_time
    
for index,match in matches.iterrows():
    
    row = []
    date = match[0]
    homeTeam = match[1]
    awayTeam = match[2]
    oddsHome = match[3]
    oddsDraw = match[4]
    oddsAway = match[5]
    homeScore = match[6]
    awayScore = match[7]
    FTR = match[8]
    HAS = teams_data[homeTeam][0]
    HDS = teams_data[homeTeam][1]
    AAS = teams_data[awayTeam][0]
    ADS = teams_data[awayTeam][1]
    
    homeRaiting, awayRaiting = style_data[homeTeam][2], style_data[awayTeam][3]
    
    HSOH = ELO_rating(homeRaiting,awayRaiting,homeScore,awayScore,'h')
    ASOH = ELO_rating(homeRaiting,awayRaiting,homeScore,awayScore,'a')
    
    style_data[homeTeam][2], style_data[awayTeam][3] = HSOH, ASOH
    
    HxG, HxGa, Hppda = find_xG(homeTeam,date)
    AxG, AxGa, Appda = find_xG(awayTeam,date)
    
    home_manager, away_manager = managers_data[homeTeam], managers_data[awayTeam]
    HMtime, AMtime = managerTime(date,home_manager),managerTime(date,away_manager)
    
    row = [HAS,HDS,HSOH,HxG,HxGa,Hppda,HMtime,AAS,ADS,ASOH,AxG,AxGa,Appda,AMtime,oddsHome,oddsDraw,oddsAway,FTR,homeTeam,awayTeam]
    
    input_data.loc[index] = row
    

    
    
    
                
    
    

input_data.to_csv('X.csv', index = False)    

input_data.head()
```
